# BAR CHART

```python
fig, ax = plt.subplots(1, 2, figsize=(17, 6))
male_data = healthdf[healthdf['Gender'] == 'Male']
ax[0].bar(x=male_data["Stress Level"], height=male_data["Sleep Duration"], color='orange')
ax[0].set_title('Distribution of Stress Level for Males', fontsize=17, fontweight="bold")
ax[0].set_xlabel("Stress Level", fontsize=15, fontweight="bold")
ax[0].set_ylabel("Sleep Duration", fontsize=15, fontweight="bold")
ax[0].set_xticks(healthdf["Stress Level"])
ax[0].set_xticklabels(healthdf["Stress Level"], fontsize=13)

female_data = healthdf[healthdf['Gender'] == 'Female']
ax[1].bar(x=female_data["Stress Level"], height=female_data["Sleep Duration"], color='blue')
ax[1].set_title('Distribution of Stress Level for Females', fontsize=17, fontweight="bold")
ax[1].set_xlabel("Stress Level", fontsize=15, fontweight="bold")
ax[1].set_ylabel("Sleep Duration", fontsize=15, fontweight="bold")
ax[1].set_xticks(healthdf["Stress Level"])
ax[1].set_xticklabels(healthdf["Stress Level"], fontsize=13)

plt.subplots_adjust(hspace=0.5)
plt.tight_layout()
plt.show()
```

```python
max_stress_level = healthdf.loc[healthdf["Sleep Duration"].idxmax()]["Stress Level"]
print(max_stress_level)
```

# SCATTER PLOT

```python
fig, ax = plt.subplots(figsize = (15,6))
Young_age = (healthdf['Age'] >= 20) & (healthdf['Age'] <= 35)
Middle_age = (healthdf['Age'] >= 36) & (healthdf['Age'] <= 50)
Old_age = (healthdf['Age'] >= 51) & (healthdf['Age'] <= 70)

ax.scatter(healthdf[Young_age]['Occupation'], healthdf[Young_age]['Quality of Sleep'],
           label='Quality of Sleep_Youth', color='darkblue', alpha=1,s=100)

ax.scatter(healthdf[Middle_age]['Occupation'], healthdf[Middle_age]['Quality of Sleep'],
           label='Quality of Sleep_Middle_Aged', color='red', alpha=1,s=100)

ax.scatter(healthdf[Old_age]['Occupation'], healthdf[Old_age]['Quality of Sleep'], 
           label='Quality of Sleep_Seniors', color='orange',alpha=1,s=100)

ax.set_xticks(range(len(healthdf['Occupation'].unique())))
ax.set_xticklabels(healthdf['Occupation'].unique(), rotation=90,fontsize=13)

ax.set_xlabel('Occupation',fontsize=15,fontweight="bold")
ax.set_ylabel('Quality of Sleep',fontsize=15,fontweight="bold")
ax.set_title('Occupational Impact on Quality of Sleep Across Different Age Ranges',fontsize=17,fontweight="bold")
ax.legend()
plt.show()
```

# HEAT MAP

```python
numeric_columns = healthdf.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = healthdf[numeric_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='PiYG', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()

```
